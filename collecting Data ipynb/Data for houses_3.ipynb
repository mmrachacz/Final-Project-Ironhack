```python
%%time
descript3 = []
price_area_rooms3 = []
loc3 = []
land3 = []
provider3 = []

for i in range(len(pages3)):
    soup = BeautifulSoup(pages3[i].content, "html.parser")
    house_info3 = soup.select("div.SearchList-22b2e")
    
    for j in range(len(house_info3)):
        
        descript3.append(house_info3[j].select('div:nth-child(1) h2')) # Offerdescription
        price_area_rooms3.append(house_info3[j].select('div:nth-child(1) div.FactsSection-52a7d div.FactsMain-bb891 div.KeyFacts-efbce div')) # price, area, rooms
        provider3.append(house_info3[j].select('div.SearchList-22b2e div.ProviderName-b71b2')) # provider
        loc3.append(house_info3[j].select('div.FactsSection-52a7d > div.FactsMain-bb891 > div:nth-child(2) > div.estateFacts-f11b0 > div:nth-child(1) > span'))
        land3.append(house_info3[j].select('div.FactsSection-52a7d > div.FactsMain-bb891 > div:nth-child(2) > div.estateFacts-f11b0 > div:nth-child(2) > span'))
```

```python
house_info3
```

```python
# flattening the list of lists
price_area_rooms_flat3 = reduce(operator.add, price_area_rooms3)
descript_flat3 = reduce(operator.add, descript3)
provider_flat3 = reduce(operator.add, provider3)
loc_flat3 = reduce(operator.add, loc3)
land_flat3 = reduce(operator.add, land3)
```

```python
# cleaning the values
price_area_rooms_flat_clean3 = [re.sub('<div data-test=|"price">|"area">|"rooms">|</div>', '', str(w)) for w in price_area_rooms_flat3]
descript_flat_clean3 = [re.sub('<h2>|</h2>', '', str(w)) for w in descript_flat3]
provider_flat_clean3 = [re.sub('<div class="ProviderName-b71b2 providerName-779ca"><span>|</span></div>', '', str(w)) for w in provider_flat3]
loc_flat_clean3 = [re.sub('<span>|</span>', '', str(w)) for w in loc_flat3]
land_flat_clean3 = [re.sub('<span>|</span>', '', str(w)) for w in land_flat3]
```

```python
# seperating values into corresponding lists (not needed for descript and provider)
price3 = price_area_rooms_flat_clean3[::3]
area3 = price_area_rooms_flat_clean3[1::3]
rooms3 = price_area_rooms_flat_clean3[2::3]
```

```python
print(len(price3))
print(len(area3))
print(len(rooms3))
print(len(loc_flat_clean3))
print(len(land_flat_clean3))
print(len(descript_flat_clean3))
print(len(provider_flat_clean3))
```

```python
houses3 = pd.DataFrame({"location":loc_flat_clean3,                       
                       "size":area3,
                       "rooms":rooms3,
                       "land_area":land_flat_clean3, 
                       "description":descript_flat_clean3,
                       "seller":provider_flat_clean3,
                       "price":price3,
                      })
houses3
```
