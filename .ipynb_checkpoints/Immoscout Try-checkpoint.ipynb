```python
list = soup.select('div.SearchList-22b2e li div div div')
```

```python
list
```

```python
import re
list = [re.sub('<div data-test=', '', str(w)) for w in list]
list = [re.sub('</div>', '', str(w)) for w in list]
list = [re.sub('"price">', '', str(w)) for w in list]
list = [re.sub('"area">', '', str(w)) for w in list]
list = [re.sub('"rooms">', '', str(w)) for w in list]
```

```python
print(list)
```

```python
price = []
area = []
rooms = []

for i in list:
    if 'm' in i:
        area.append(i)
    if '€' in i:
        price.append(i)
    if 'Zi' in i:
        rooms.append(i)
```

```python
print(price)
print(area)
print(rooms)
```

```python
houses = pd.DataFrame({"price":price,
                       "area":area,
                       "rooms":rooms})
houses
```
