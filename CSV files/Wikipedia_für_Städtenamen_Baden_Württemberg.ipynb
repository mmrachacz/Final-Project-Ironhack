```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
from time import sleep
import time
from random import randint
import re
import operator
from functools import reduce
```

```python
url = 'https://de.wikipedia.org/wiki/Liste_der_gr%C3%B6%C3%9Ften_St%C3%A4dte_in_Baden-W%C3%BCrttemberg'
```

```python
response = requests.get(url)
response.status_code
```

```python
soup = BeautifulSoup(response.content, "html.parser")
```

```python
print(soup.prettify())
```

```python
#mw-content-text > div.mw-parser-output > table > tbody > tr:nth-child(1) > td:nth-child(2)
soup.select('td:nth-child(2)')#[0].get_text()
```

```python
city = []

num_iter = len(soup.select('td:nth-child(2)'))

info = soup.select('td:nth-child(2)')

for i in range(num_iter):
    city.append(info[i].get_text())
    
city
```

```python
city_df = pd.DataFrame(city)
```

```python
city_df.to_csv('City_Names_Baden_WRTMBRG_List', index=False)
```
